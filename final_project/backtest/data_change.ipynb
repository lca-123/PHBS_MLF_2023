```python
import pandas as pd 
import pickle
```

```python
df_factor = pd.read_parquet('../Data/因子数据_月频_完整.parquet')
df_price = pd.read_parquet('../Data/价格数据_完整.parquet')
df_pool = pd.read_parquet('../Data/股票池数据_完整.parquet')
df_benchmark = pd.read_csv('../Data/index_close.csv')
```

```python
df_factor.rename(columns={'交易日期': 'date','股票代码':'asset','行业':'group'},inplace=True)
df_factor.set_index(['date','asset'],inplace = True)

df_group = df_factor[['group']]

df_factor = df_factor[['momentum240_20','turnover240_20']]

df_price.rename(columns={'交易日期': 'date','股票代码':'asset','日收盘价':'price'},inplace=True)
df_price.set_index(['date','asset'],inplace= True)

df_pool.rename(columns={'交易日期': 'date','股票代码':'asset'},inplace=True)
df_pool.set_index(['date','asset'],inplace=True)
dict_df_pool ={'000905': df_pool}

df_benchmark['date'] = pd.to_datetime(df_benchmark['date'])
df_benchmark.set_index('date',inplace=True)
df_benchmark = df_benchmark[['000905']]
dict_df_benchmark ={'000905': df_benchmark}
```
